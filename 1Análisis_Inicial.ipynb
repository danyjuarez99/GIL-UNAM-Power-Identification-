```python
#Carga los df de cada idioma
pd.set_option('display.max_colwidth', 50)
es = pd.read_csv(r"C:\Users\madja\Desktop\Servicio Social\Power Identification in Parliamentary Debates\es-train.tsv",sep = '\t')
ct = pd.read_csv(r"C:\Users\madja\Desktop\Servicio Social\Power Identification in Parliamentary Debates\es-ct-train.tsv",sep = '\t')
ga = pd.read_csv(r"C:\Users\madja\Desktop\Servicio Social\Power Identification in Parliamentary Debates\es-ga-train.tsv",sep = '\t')
pv = pd.read_csv(r"C:\Users\madja\Desktop\Servicio Social\Power Identification in Parliamentary Debates\es-pv-train.tsv",sep = '\t')
es
```

```python
datasets = {'es':es,'ct':ct,'ga':ga,'pv':pv}
for name, df in datasets.items():
    print(f"El dataset {name} tiene {len(df)} discursos.")
```

```python
#Visualiza el discurso del idioma df
def display(index,df):
    fila = pd.DataFrame(df.loc[index])
    fila = fila.drop(['id','speaker','sex','text_en'])
    return fila
```

```python
pd.set_option('display.max_colwidth', 25000)
display(16,es)
```

```python
import random
def random_speech():
    keys_list = list(datasets.keys())
    random_key = random.choice(keys_list)
    random_set = datasets[(random_key)]
    index = random.randint(1,len(random_set)+1)
    print(f'Se eligió al azar la fila {index} del dataset {random_key}')
    return display(index,random_set)
```
